```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

df = pd.read_csv(filepath, names=headers)
```

```python
df.head()
```

```python
df.replace("?", np.nan, inplace = True)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
##Replace "NaN" with the mean value
```

```python
avd_nl = df["normalized-losses"].astype("float").mean(axis=0)
print("ava: ", avd_nl)
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avd_nl)
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
df["bore"] = df["bore"].replace(np.nan, avg_bore)
avg_hp = df["horsepower"].astype("float").mean(axis=0)
df["horsepower"] = df["horsepower"].replace(np.nan, avg_hp)
```

```python
## 
```

```python
## Replace "NaN" with the frequent value
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, 'four')
```

```python
df['num-of-doors'].value_counts()
```

```python
## Drop the whole row:
```

```python
df.dropna(subset = "price", inplace = True)
```

```python
df.reset_index(drop=True, inplace = True)
```

```python
df
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
# data standardization - prepares the data for comparing two similar data formats
```

```python
df['city-L/100'] = 235 / df['city-mpg']
```

```python
df.head()
```

```python
# df['highway-mpg'] = 235/df['highway-mpg']
df.rename(columns={'highway - L/100': "Highway-L/100km"}, inplace = True)
df.head()
```

```python
# data Normalization -  transforming values of several variables into a similar range
```

```python
df["height"] = df["height"]/df["height"].max()
df["height"].head()
```

```python
df["length"] = df["length"]/df["length"].max()
df["length"].head()
```

```python
# Bining: transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis
```

```python
df["horsepower"] = df["horsepower"].astype("int")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_name = ["Low", "Medium", "High"]
df["HP-binned"] = pd.cut(df["horsepower"], bins, labels=group_name, include_lowest = True)
df[["HP-binned", "horsepower"]].tail(20)
```

```python
df["HP-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_name, df["HP-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Indicator Variables (Dummy variable):   is a numerical variable used to label categories.
#                       - use categorical variables for regression analysis
```

```python
dv_1 = pd.get_dummies(df["fuel-type"])
df["fuel-type"].tail(50)
df["fuel-type"].value_counts()
```
