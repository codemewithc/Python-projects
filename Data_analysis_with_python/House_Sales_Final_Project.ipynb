```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset (update path if your file name is different)
df = pd.read_csv('kc_house_data.csv')

# Preview the data
print(df.head())
```

```python
# data wrangling
```

```python
# Drop the 'id' and 'Unnamed: 0' columns if they exist
# errors='ignore' ensures it doesn't crash if the column is already gone
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True, errors='ignore')

# Check for missing values in bedrooms and bathrooms (common in this dataset)
print(df.isnull().sum())

# Replace missing values with the mean
df['bedrooms'].replace(np.nan, df['bedrooms'].mean(), inplace=True)
df['bathrooms'].replace(np.nan, df['bathrooms'].mean(), inplace=True)

print("Data cleaning complete.")
```

```python
#Exploratory Data Analysis
```

```python
# Count the number of houses with unique floor values
floor_counts = df['floors'].value_counts().to_frame()
print(floor_counts)

# Use a boxplot to see if houses with a waterfront view have more price outliers
sns.boxplot(x='waterfront', y='price', data=df)
plt.title("Price Distribution: No Waterfront (0) vs. Waterfront (1)")
plt.show()

# Use a regression plot to see the correlation between 'sqft_above' and 'price'
sns.regplot(x='sqft_above', y='price', data=df)
plt.title("Correlation: Price vs. Square Footage (Above)")
plt.show()
```

```python
#Model Development
```

```python
# Simple Linear Regression (Using sqft_living)
X = df[['sqft_living']]
Y = df['price']
lm = LinearRegression()
lm.fit(X, Y)
print(f"Simple Linear Regression R^2: {lm.score(X, Y):.4f}")

# Advanced Modeling: Creating a Pipeline
# This scales the data, creates polynomial features, and runs a Ridge regression
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", 
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]
X_multi = df[features]

Input = [('scale', StandardScaler()), 
         ('polynomial', PolynomialFeatures(include_bias=False)), 
         ('model', LinearRegression())]

pipe = Pipeline(Input)
pipe.fit(X_multi, Y)
print(f"Pipeline R^2 (Multi-feature): {pipe.score(X_multi, Y):.4f}")
```

```python
#Model Evaluation
```

```python
# Split the data
x_train, x_test, y_train, y_test = train_test_split(X_multi, Y, test_size=0.15, random_state=1)

# Ridge Regression with Alpha set to 0.1
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)
test_score = RidgeModel.score(x_test, y_test)

print(f"Ridge Regression R^2 on Test Data: {test_score:.4f}")
```
