```python
import pandas as pd
import numpy as np
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
```

```python
df = pd.read_csv(file_path, header = None)
df.head(5)
```

```python
headers = ['Manufacturer', 'Category', 'screen', 'GPU', 'Os', 'CPU_Core', 'Screen_size_cm', 'CPU_Frequency', 'Ram_GB', 'Storage_GB_SSD', 'Weight_Kg', 'Price']
df.columns = headers
df.head(10)
```

```python
df.dtypes
```

```python
df.info()
```

```python
df = df.replace("?", np.nan, inplace = True)
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_Weight_Kg = df["Weight_Kg"].astype("float").mean(axis=0)
print("Average price is: ", avg_Weight_Kg)
df["Weight_Kg"] = df["Weight_Kg"].replace(np.nan, avg_Weight_Kg)
df.head(16)
```

```python
avg_Screen_size_cm = df["Screen_size_cm"].value_counts().idxmax()
print("Average Screen_size_cm is: ", avg_Screen_size_cm)
df["Screen_size_cm"] = df["Screen_size_cm"].replace(np.nan, avg_Screen_size_cm)
df.tail(10)
```

```python
df["Screen_size_cm"] = df["Screen_size_cm"].astype("float")
df["Weight_Kg"] = df["Weight_Kg"].astype("float")
```

```python
df["weight_in_lbs"] = df["Weight_Kg"] * 2.2
df["screen_size_inch"] = df["Screen_size_cm"] / 2.54
```

```python
df["CPU_Frequency"] = df["CPU_Frequency"] / df["CPU_Frequency"].max()
```

```python
bins = np.linspace(df["Price"].min(), df["Price"].max(), 4)
group_names = ["Low", "Medium", "High"]
df["Price_binning"] = pd.cut(df["Price"], bins, labels=group_names, include_lowest = True)
df.tail()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

pyplot.bar(group_names, df["Price_binning"].value_counts())
plt.pyplot.xlabel("Price")
plt.pyplot.ylabel("Count") 
plt.pyplot.title("Price Binning")
```
