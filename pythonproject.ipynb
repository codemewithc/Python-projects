```python
# Exercise 1
# Extract the required GDP data from the given URL using Web Scraping.
URL = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
# async def download(url, filename):
#    response = await pyfetch(url)
#    if response.status == 200:
#         with open ("sample.csv") as s:
#             s.write()
URL = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
table = pd.read_html(URL)
df = table[3]

df.columns = range(df.shape[1])

df = df[[0,2]]

df = df.iloc[1:11,:]

df.columns = ['country', 'GDP(in mil)']

df
```

```python
# Exercise 2
# Modify the GDP column of the DataFrame, converting the value available in Million USD to Billion USD. 
# Use the round() method of Numpy library to round the value to 2 decimal places. Modify the header of the 
# DataFrame to GDP (Billion USD).

df['GDP(in mil)'] = df['GDP(in mil)'].astype(int)

df[['GDP(in mil)']] = df[['GDP(in mil)']]/1000

df[['GDP(in mil)']] = np.round(df[['GDP(in mil)']], 2)

df.columns = ["country", "GDP(in bil)"]

df
```

```python
# Exercise 3
# Load the DataFrame to the CSV file named "Largest_economies.csv"

df.to_csv('./Largest_economies.csv')
```
